<h3> Definição dos Bairros de cada estabelecimento </h3>

```python
## Colocando em qual bairro cada estabelecimento pertence

## Abre o arquivo de shapefile que contem as informações dos bairros
shp = shapefile.Reader("Toronto_Zoning_per_Neighbourhood.shp")
all_shapes = shp.shapes()
all_records = shp.records()     

## Cria uma lista com todos os shapes 'prontos'
shapes_ready = []
for i in all_shapes:
    shapes_ready.append(shape(i))

## Função que dado uma (longitude, latitude), retorna o bairro de toronto que se encontra
def get_neighbourhood(point_to_check):
    ## Cria o ponto
    point = Point(point_to_check)
    ## Percorre a lista dos shapes
    for i in range(0, len(all_shapes)):
        ## Pega o shape
        boundary = shapes_ready[i]
        ## Verifica se o ponto está dentro
        if point.within(boundary):
            ## Se estiver, pega o nome do bairro no all_records
            name = all_records[i][1]
            return name
    return False

## Cria uma nova coluna no dataframe para colocar o bairro
df_x_train["neighbourhood"] = df_x_train.apply(lambda row: get_neighbourhood( (row["longitude"], row["latitude"]) ), axis=1)
```

<h3> Separando os atributos - Colocando eles em colunas </h3>

```python
##### PARTE 1 - SEPARAÇÃO DE TODOS OS ATRIBUTOS PRESENTES NOS DADOS ############################
## Pega a lista de todos os atributos
attrs = df_x_train['attributes']

## Variaveis para armazenar as chaves e uma lista dos valores possiveis
values = {}

## Percorre a lista dos atributos
for i in attrs:
    ## Se for nan (nao tiver atributo), volta pro topo
    if str(i) == "nan":
        continue
    else:
        ## Transforma o objeto em um dicionário
        aux = ast.literal_eval(i)
        
        ## Percorre Esse dicionário
        for j in aux:
            ## Transforma o valor (se der, vai transformar para um dicionário também)
            aux_2 = ast.literal_eval(aux[j])
            ## Se tiver valor nulo, apenas pula e volta para o inicio do loop
            if type(aux_2) == type(None):
                continue
            ## Verifica se a chave que está percorrendo já foi identificada
            if j in values:
                ## Se for um dicionário dentro
                if type(aux_2) is dict:
                    ## Pecorre esse dicionário
                    for k in aux_2:
                        ## Verifica se a chave já foi identificada
                        ## Senão foi, adiciona
                        if k not in values[j]:
                            values[j][k] = ""
                            continue
                                        
            ############## Caso a chave não tenha sido identificada
            else:
                ## Verfica se é um dicionário ou um valor 'normal'
                if type(aux_2) is dict:
                    ## Cria uma variavel para colocar internamente na armazenagem dos valores
                    dict_interno = {}
                    ## Percorre esse dicionário interno
                    for k in aux_2:
                        ## Preenche o dicionário interno
                        dict_interno[k] = ""
                    ## Faz o dicionário externo receber o dicionário interno criado
                    values[str(j)] = dict_interno
                ## Caso seja um valor normal
                else:
                    ## Adiciona nas variaveis, uma lista com aquele valor
                    values[str(j)] = ""
####################################################################

##### PARTE 2 - CRIANDO COLUNAS PARA CADA ATRIBUTOS ###########
## Se caso algum estabelecimento não tiver os dados, será colocado como FALSO por padrão

## Função que recebe um dicionário (x), uma chave (i), e uma chave interna (k)
def get_value(x, i, k = False):
    ## Se x for nulo, então retorna False
    if str(x) == 'nan':
        return False
    
    ## Se não houver chave interna (ou seja, x[i] não é um dicionário)
    if not k:
        try:
            ## Transforma o X em um dicionario e retorna o valor de x[i]
            dic = ast.literal_eval(x)
            return dic[i]
        except Exception as e:
            return False
    ## Caso x[i] for um dicionário
    else:
        ## Tenta transforma e retorna o valor de x[i][k]
        try:
            dic = ast.literal_eval(x)
            dic = ast.literal_eval(dic[i])
            return dic[k]
        except Exception as e:
            return False

## Percorre todas as chaves encontradas anteriormente        
for i in values.keys():
    ## Verifica se os dados são de um dicionário
    if type(values[i]) == dict:
        ## Percorre o dicionário, se for o caso, e chama a função para retornar o valor
        for k in values[i].keys():
             df_x_train[str(i) + "_" + str(k)] = df_x_train["attributes"].apply(lambda x: get_value(x, i, k))
    ## Cria a coluna com o atributo e chama a função para retornar o valor adequado
    else:    
        df_x_train[i] = df_x_train["attributes"].apply(lambda x: get_value(x, i))
        
####################################################################
```

<h3> Análise da faixa de Preços </h3>

```python
## Função para mostrar um gráfico de pizza com os preços dos estabelecimentos
def show_range_prices(data):
    counter = Counter(data)
    counter = counter.most_common()
    counter = dict(counter)
    fig = plt.figure()
    ax = fig.add_axes( [0,0,1,1])
    ax.axis('equal')
    langs = list(counter.keys())
    students =  list(counter.values()) 
    ax.pie(students, labels = langs,autopct='%1.2f%%')
    plt.show()
```

<h3> Análise das frequencias entre Destaques x Não Destaques </h3>

```python
## Função para realizar a contagem de uma lista e mostrar os 5 elementos mais comuns e os 5 mais incomuns
def count_things(df, column):
    ## Realiza a contagem
    counter = Counter(df[column])

    ## Pega os 5 menos comuns
    least_common = counter.most_common()[:-6:-1]
    ## Pega os 5 mais comuns
    most_common = counter.most_common(5)

    print(f"-----> {column} mais comuns - Combinações <-----")
    for i in most_common:
        print(f"{i}")
    print("\n\n")
    print(f"-----> {column} menos comuns  - Combinações<-----")
    for i in least_common:
        print(f"{i}")
        
print("#-#-#-#-#-#-# LUGARES DE DESTAQUE #-#-#-#-#-#-#")
count_things(df_x_destaques, 'categories')
print("\n #-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-##-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-# \n")
print("#-#-#-#-#-#-# LUGARES DE NÃO DESTAQUE #-#-#-#-#-#-#")
count_things(df_x_nao_destaques, 'categories')
```

<h2> Análise das reviews - Texto </h2>

```python
## Funções usadas para gerar as nuvens de palvras
def generate_most_used_worlds(df):
    ## Pega a coluna de texto apenas dos 100 primeiros elementos
    df_x = df[['text']].head(100)
    
    ## Cria uma string com todos os comentários
    all_words = df_x.text.sum()
    ## Cria uma função para retirar as palavras desnecessárias
    def remove_unuseful_words(tokens):
        nlp = spacy.load("en_core_web_sm")
        doc = nlp(tokens)
        ## Só coloca na lista de retorno as palavras que são classificadas como adjetivo
        return [str(t) for t in doc if (t.pos_ == 'ADJ')]
    ## Conta quantas vezes cada palavra se repetiu
    counter = Counter(remove_unuseful_words(all_words))
    ## Mostra essas quantidades em um gráfico
    freq_df = pd.DataFrame.from_records(counter.most_common(20), columns=['token', 'count'])
    freq_df.plot(kind='bar', x='token')
    return counter

def wordcloud(counter):
    wc = WordCloud(width=1200, height=800,
                   background_color="white",
                   max_words=200)
    wc.generate_from_frequencies(counter)
    
    fig = plt.figure(figsize=(6,4))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
```

```python
## Faz o diagrama de palavras mais usadas nos reviews nos lugares que são destaques
print("Palavras mais usadas em reviews em lugares de destaque: \n")
## Pega apenas os reviews que tiveram mais de 20 'useful'
df_x_destaques_full_more_20 = df_x_destaques_review[df_x_destaques_review['useful'] > 20]
counter = generate_most_used_worlds(df_x_destaques_full_more_20)
wordcloud(counter)
```

```python
## Faz o diagrama de palavras mais usadas nos reviews nos lugares que não são destaques
print("Palavras mais usadas em reviews em lugares que não são destaque: \n")
## Pega apenas aqueles reviews que tem mais de 20 'useful'
df_x_nao_destaques_full_more_20 = df_x_nao_destaques_review[df_x_nao_destaques_review['useful'] > 20]
counter = generate_most_used_worlds(df_x_nao_destaques_full_more_20)
wordcloud(counter)
```
