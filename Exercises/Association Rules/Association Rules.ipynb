```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
import numpy as np

#!pip install mlxtend
```

<h3> Exercicio 1 </h3>

```python
#Lista de transações
transactions  =   [['tree','cup', 'paper'],
                         ['book','tomato','pen'],
                         ['pen', 'book', 'tree'],
                         ['tomato', 'pen', 'cup'],
                         ['tree', 'paper', 'book'],
                         ['pen', 'book']]
te = TransactionEncoder()
encoded = te.fit(transactions).transform(transactions)
df = pd.DataFrame(encoded, columns=te.columns_)
#Calcula os conjuntos de itens (itemsets) com suporte >= 0,33
frequent_itemsets = apriori(df, min_support=0.33, use_colnames=True)
frequent_itemsets
```

```python
## Para pegar apenas aqueles conjuntos que tem exatamente um suporte de 33%
frequent_itemsets[frequent_itemsets["support"] == 1/3]
```

```python
transactions  =   [['1', '3', '5', '7', '8'],
                         ['3','4','5', '7'],
                         ['2', '3', '5', '6'],
                         ['2', '4', '5', '8', '9'],
                         ['2', '3', '5', '11'],
                         ['1', '2', '3', '7', '9']]

#Recurso da biblioteca mlxtend para codificar transações no padrão numpy array booleano - cada coluna é um item 
#e linhas indicam se esse item existe em uma transação.
te = TransactionEncoder()
encoded = te.fit(transactions).transform(transactions)
dfExemplo = pd.DataFrame(encoded, columns=te.columns_)
print(dfExemplo)

#Calcula os conjuntos de itens (itemsets) com suporte >= 0,33
frequent_itemsets = apriori(dfExemplo, min_support=0.0001, use_colnames=True)
frequent_itemsets
```

```python
#Regras com o valor de confiança mínimo escolhido
dfRegras= association_rules(frequent_itemsets, metric='confidence', min_threshold=0.73)
dfRegras
```

<h3> Exercicio 3 </h3>

```python
## Abrindo o arquivo de supermercado.csv
data = open("supermercado.csv")
linhas = data.readlines()
## Le esse arquivo, e constroi uma lista de listas
transactions = []
for i in linhas:
    i = i.replace("\n", "")
    aux = i.split(',')
    aux = sorted(set(aux))    
    transactions.append(aux)
```

```python
## Usa o algoritmo APRIORI com suporte minimo de 0.001
te = TransactionEncoder()
encoded = te.fit(transactions).transform(transactions)
df = pd.DataFrame(encoded, columns=te.columns_)
frequent_itemsets = apriori(df, min_support=0.001, use_colnames=True)
frequent_itemsets
```

```python
## Regras com o valor de confiança mínimo 25%
dfRegras = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.25)

## Pega apenas as regras que tem como consequente 'bottled beer'
dfRegras = dfRegras[dfRegras['consequents'] == {"bottled beer"}]
dfRegras
```

Parte a

```python
df_reviews = pd.read_csv("reviewsTrainToronto.csv")
df_X = pd.read_csv("X_trainToronto.csv")
```

```python
## Separa os lugares que o um usuário fez pelo menos dois reviews
dfGruped = df_reviews.groupby(by=["user_id", "business_id"]).count()
dfGruped = dfGruped[dfGruped['useful'] > 1]
## Conta quantas pessoas realizaram fizeram mais de dois reviews
dfGruped['qtd'] = 1
dfGruped = dfGruped.reset_index()
dfGruped = dfGruped[['business_id', 'qtd']]
dfGruped = dfGruped.groupby(by=["business_id"]).sum()
dfGruped = dfGruped.reset_index()
```

```python
## Faz o Merge com os dados dos estabelecimentos
dfMerged = dfGruped.merge(df_X, left_on='business_id', right_on='business_id')
dfMerged = dfMerged[['qtd', 'categories', 'destaque']]
dfMerged
```

```python
## Cria a lista de transações, com o ultimo elemento sendo Sim ou não, referente a ser destaque ou não.
lista_categorias = dfMerged['categories']
lista_destaque = dfMerged['destaque']
transactions = []
cont = 0
for i in lista_categorias:
    aux = str(i).split(', ')
    if lista_destaque[cont] == 0:
        aux.append("Não")
    else:
        aux.append("Sim")
    transactions.append(aux)
    cont += 1
```

```python
##Transforma as transações em um dataframe
te = TransactionEncoder()
encoded = te.fit(transactions).transform(transactions)
df = pd.DataFrame(encoded, columns=te.columns_)
#Calcula os conjuntos de itens (itemsets) com suporte >= 0,001
frequent_itemsets = apriori(df, min_support=0.0015, use_colnames=True)
frequent_itemsets
```

Parte b

```python
#Regras com o valor de confiança mínimo escolhido de 30%
dfRegras= association_rules(frequent_itemsets, metric='confidence', min_threshold=0.3)
## Filtra para pegar apenas regras que impliquem em sim ou não
dfRegras = dfRegras[ (dfRegras['consequents'] == {"Não"}) | (dfRegras['consequents'] == {"Sim"})]
dfRegras
```

```python
#Regras com o valor de confiança mínimo escolhido de 90%
dfRegras= association_rules(frequent_itemsets, metric='confidence', min_threshold=0.9)
dfRegras = dfRegras[ (dfRegras['consequents'] == {"Não"}) | (dfRegras['consequents'] == {"Sim"})]
dfRegras
```

```python
df_reviews = pd.read_csv("reviewsTrainToronto.csv")
df_X = pd.read_csv("X_trainToronto.csv")
```

```python
def find_more_pleope2_reviews(df_in, df_reviews_in):
    df = df_in.copy()
    df_reviews = df_reviews_in.copy()
    ## Separa os lugares que o um usuário fez pelo menos dois reviews
    dfGruped = df_reviews.groupby(by=["user_id", "business_id"]).count()
    dfGruped = dfGruped[dfGruped['useful'] > 1]
    ## Conta quantas pessoas realizaram fizeram mais de dois reviews
    dfGruped['qtd'] = 1
    dfGruped = dfGruped.reset_index()
    dfGruped = dfGruped[['business_id', 'qtd']]
    dfGruped = dfGruped.groupby(by=["business_id"]).sum()
    dfGruped = dfGruped.reset_index()
    ## Faz o Merge com os dados dos estabelecimentos
    dfMerged = df_X.merge(dfGruped, left_on='business_id', right_on='business_id', how='left')
    dfMerged['qtd'] = dfMerged['qtd'].fillna(-1)
    return dfMerged

def find_features_attr_reviews(df_in, df_reviews_in):
    df = df_in.copy()
    df_reviews = df_reviews_in.copy()
    df_use = find_more_pleope2_reviews(df, df_reviews)
    dfMerged = df_use[df_use['qtd'] != -1]
    
    dfMerged = dfMerged[['qtd', 'categories', 'destaque']]
    
    lista_categorias = list(dfMerged['categories'])
    lista_destaque = list(dfMerged['destaque'])
    transactions = []
    cont = 0
    for i in lista_categorias:
        aux = str(i).split(', ')
        if lista_destaque[cont] == 0:
            aux.append("Não")
        else:
            aux.append("Sim")
        transactions.append(aux)
        cont += 1

    te = TransactionEncoder()
    encoded = te.fit(transactions).transform(transactions)
    df = pd.DataFrame(encoded, columns=te.columns_)
    #Calcula os conjuntos de itens (itemsets) com suporte >= 0,001
    frequent_itemsets = apriori(df, min_support=0.001, use_colnames=True)
    dfRegras= association_rules(frequent_itemsets, metric='confidence', min_threshold=1)
    dfRegras = dfRegras[ (dfRegras['consequents'] == {"Não"}) | (dfRegras['consequents'] == {"Sim"})]
    
    itens = []
    for i in dfRegras['antecedents']:
        itens.append(list(i))
    return itens
```
